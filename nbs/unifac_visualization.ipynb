```python
from deep_gamma import DATA_PATH
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from rdkit import Chem
from IPython.display import display
```

```python
unifac_df = pd.read_csv(DATA_PATH  / "07_model_output" / "aspen_unifac.csv", index_col=0)
aspen_df = pd.read_csv(DATA_PATH / "01_raw" / "aspen_data.csv")
```

```python
df = pd.concat([unifac_df, aspen_df[["ln_gamma_1", "ln_gamma_2"]]], axis=1)
for i in [1,2]:
    df[f"ln_gamma_{i}_error"] = (df[f"ln_gamma_{i}_pred"] - df[f"ln_gamma_{i}"]).abs()
filtered_df = df.dropna().sort_values("ln_gamma_1_error", ascending=True)
```

```python
j = 0
colors = ["#025b66", "#b7001e"]
for ind, group in filtered_df.groupby(["smiles_1", "smiles_2", "TRange"]):
    if j == 5:
        break
    img = Chem.Draw.MolsToGridImage(
        [Chem.MolFromSmiles(smiles) for smiles in ind[:2]]
    )
    display(img)
    fig, ax = plt.subplots(1)
    for i in [1,2]:
        group.plot(x="x1", y=[f"ln_gamma_{i}_pred"], ax=ax, kind="scatter", color=colors[0])
        group.plot(x="x1", y=[f"ln_gamma_{i}"], ax=ax, color=colors[1])
    ax.set_label("x")
    ax.set_ylabel(r"$\ln \gamma$")
    unifac = mpatches.Patch(label=r"UNIFAC", color=colors[0])
    experiment = mpatches.Patch(label=r"Aspen (experimental)", color=colors[1])
    ax.legend(handles=[unifac, experiment])
    display(fig)
    display(group)
    j+=1
```
