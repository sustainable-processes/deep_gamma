```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_df = pd.read_csv("../data/05_model_input/cosmo/train.csv")
test_df = pd.read_csv("../data/05_model_input/cosmo/test_indp.csv")
```

```python
train_df["split"] = "Train"
test_df["split"] = "Test"
all_data = pd.concat([train_df, test_df]).reset_index()
```

```python
sns.kdeplot(
    data=all_data, x="ln_gamma_1", hue="split",common_norm=False
)
```

```python
fig, ax = plt.subplots()
sns.kdeplot(
    data=all_data, x="ln_gamma_1", hue="split",common_norm=False, ax=ax
)
ax.set_xlim(-3, 3)
```
